```python
#--task1  (lesson9)
#-- oracle: https://www.hackerrank.com/challenges/the-report/problem

Ketty gives Eve a task to generate a report containing three columns: Name, Grade and Mark.
Ketty doesn't want the NAMES of those students who received a grade lower than 8. 
The report must be in descending order by grade -- i.e. higher grades are entered first. 
If there is more than one student with the same grade (8-10) assigned to them, 
order those particular students by their name alphabetically. 
Finally, if the grade is lower than 8, use "NULL" as their name and list them by their grades in descending order. 
If there is more than one student with the same grade (1-7) assigned to them,
order those particular students by their marks in ascending order.

Write a query to help Eve.
```

```python
Students=pd.DataFrame([[1,'Julia',88],[2,'Samantha',68],[3,'Maria',99],[4,'Skarlet',78],[5,'Ashley',63],[6,'Jain',81]],
        columns=['id','Name','Marks'])
Students
```

```python
Grades=pd.DataFrame([[1,0,9],[2,10,19],[3,20,29],[4,30,39],[5,40,49],[6,50,59],
                    [7,60,69],[8,70,79],[9,80,89],[10,90,100]],
                    
        columns=['Grade','Min_Mark','Max_Mark'])
Grades
```

```python
import sqlite3
conn = sqlite3.connect('TestDB.db')  
c = conn.cursor()

request = "CREATE TABLE df_Students(id int, Name varchar(10), 'Marks' int)"
c.execute(request)
tables = c.fetchall()

Students.to_sql('df_Students', con = conn, if_exists = 'append', index = False)
```

```python
import sqlite3
conn = sqlite3.connect('TestDB.db')  
c = conn.cursor()

request = "CREATE TABLE df_Grades(Grade int, Min_Mark int, Max_Mark int)"
c.execute(request)
tables = c.fetchall()

Grades.to_sql('df_Grades', con = conn, if_exists = 'append', index = False)
```
